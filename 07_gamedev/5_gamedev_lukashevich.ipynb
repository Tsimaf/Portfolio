## Описание проекта

Я работаю в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Мне нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. 

Передо мной данные до 2016 года. В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Описание данных**

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

**План проекта:**
- Загрузка данных и просмотр общей информации
- Подготовка данных

## Проверка гипотез

### Проверка гипотезы 1: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Создадим датафреймы для каждой категории:

```python
df_XOne = df_relevant.query('platform=="XOne"')
df_PC = df_relevant.query('platform=="PC"')
```

Необходимо составить нулевую и альтернативную гипотезу:
- Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC равны.
- Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC разные.

Далее зададим уровень значимости alpha и проведем ttest:

```python
alpha = 0.05
results = st.ttest_ind(df_XOne['user_score'], df_PC['user_score'], equal_var=False, nan_policy='omit')
results.pvalue
```

```python
if results.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

Посмотрим на средние значения для каждой платформы:

```python
print(df_XOne['user_score'].mean())
print(df_PC['user_score'].mean())
```

Полученное значение p-value больше заданного уровня значимости, а это значит, что у нас нет оснований отвергнуть нулевую гипотезу. А значит у нас нет основания говорить, что средние пользовательские рейтинги платформ Xbox One и PC различны.

### Проверка гипотезы 2: средние пользовательские рейтинги жанров Action и Sports разные

Создадим датафреймы для каждой категории:

```python
df_action = df_relevant.query('genre=="Action"')
df_sports = df_relevant.query('genre=="Sports"')
```

Составим нулевую и альтернативную гипотезу:
- Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports равны.
- Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

Проведем ttest:

```python
results_g = st.ttest_ind(df_action['user_score'], df_sports['user_score'], equal_var=False, nan_policy='omit')
results_g.pvalue
```

```python
if results_g.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

Посмотрим средние значения:

```python
print(df_action['user_score'].mean())
print(df_sports['user_score'].mean())
```

Полученное значение p-value меньше заданного уровня значимости, поэтому можно отвергнуть нулевую гипотезу. А это значит, что есть основания говорить, что средние пользовательские рейтинги жанров Action и Sports разные.

Для проверки этих гипотез мы использовали метод ttest_ind, потому что на вход нам нужно было передать выборки разного размера. Нам не нужно было изменять параметр Alternative, потому что по умолчанию он проверяет двухсторонюю гипотезу - именно ту, которую нам нужно было проверить в альтернативных гипотезах. Мы указали значение параметра nan_policy='omit', чтобы тест игнорировал пропущенные значения. Нулевая гипотеза - всегда гипотеза о равенстве.
